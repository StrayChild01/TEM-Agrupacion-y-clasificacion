```r
library(readxl)

actitud_compras <- read_excel("datasets/Actitud_Compras.xlsx", sheet="Hoja1")
actitud_compras
```

```r
casos_act_compras <- actitud_compras[,grep("X",names(actitud_compras))]
dist_compras <- round(proxy::dist(casos_act_compras, method="euclidean", upper=T, diag=TRUE),2)
as.matrix(dist_compras)
```

```r
clus_act_compras <- hclust(dist_compras, method="single")
plot(sort(clus_act_compras$height, decreasing=TRUE), ylab="distancia", xlab="Num clusters")
```

```r
library(ggplot2)
suppressPackageStartupMessages(library(dendextend))
library(dplyr)

num_clusters <- 4

dend <- clus_act_compras %>%
    as.dendrogram %>%
    set("branches_k_color", k=num_clusters) %>% #color de las ramas
    set("branches_lwd", 1.2) %>% # ancho de las ramas
    set("labels_cex", .5) %>% # tamaño de las hojas
    set("labels_colors", k=num_clusters) %>% # Colores de las etiquetas
    # set("leaves_pch", c(19,1,4)) %>% # tipo de punto en las hojas 
    # Parece que la opción anterior no funciona.
    set("leaves_cex", 0.5) #  color de las hojas
        
ggd1 <- as.ggdend(dend)
ggplot(ggd1, horiz = F, theme=theme_minimal())+
    labs(
        title = "Actitudes de compradores"
        ,subtitle = "Grupos de compradores"
        ,x = "# clusters"
        ,y = "Distancia"
    )
```

```r
clus_act_compras_w <- hclust(dist_compras, method="ward.D")
plot(sort(clus_act_compras_w$height, decreasing=TRUE), ylab="distancia", xlab="Num clusters")
```

```r
library(ggplot2)
suppressPackageStartupMessages(library(dendextend))
library(dplyr)

num_clusters <- 3

dend_w <- clus_act_compras_w %>%
    as.dendrogram %>%
    set("branches_k_color", k=num_clusters) %>% #color de las ramas
    set("branches_lwd", 1.2) %>% # ancho de las ramas    
    #set("branches_lty", .5) %>% # Dashed branches
    set("labels_cex", .5) %>% # tamaño de las hojas
    set("labels_colors", k=num_clusters) %>% # Colores de las etiquetas
    set("nodes_pch", 19) %>% # Puntos en los nodos
    set("nodes_cex", 2) %>% # tamaño del punto anterior
    # set("leaves_pch", c(19,1,4)) %>% # tipo de punto en las hojas    
    # Parece que la opción anterior no funciona.
    set("leaves_cex", 0.5) #  color de las hojas
        
ggd1 <- as.ggdend(dend_w)
ggplot(ggd1, horiz = F, theme=theme_minimal())+
    labs(
        title = "Actitudes de compradores"
        ,subtitle = "Grupos de compradores"
        ,x = "# clusters"
        ,y = "Distancia"
    )
```

```r
hist_clus_act_compras <- data.frame(cbind(1:19,clus_act_compras_w$merge, clus_act_compras_w$height ))
colnames(hist_clus_act_compras) <- c("#","ind 1", "ind 2", "distancia")
hist_clus_act_compras
```

```r
plot(clus_act_compras_w)
rect.hclust(clus_act_compras_w, k=3, border="red")
```

```r
clu_tree <- cutree(clus_act_compras_w, k=3)
table(clu_tree)
```

```r
actitud_compras <- cbind(actitud_compras,clu_tree)
actitud_compras[with(actitud_compras, order(clu_tree)),c("Caso", "edad", "clu_tree")]
```

```r
library(factoextra)
# Hello darkness my old friend...
pca_actitud_compras <- prcomp(actitud_compras[,c(2:7)], scale.=T, retx=T)

clusters <- as.factor(actitud_compras$clu_tree)

fviz_pca_ind(pca_actitud_compras,
             col.ind = clusters, # color by groups
             #palette = c("#00AFBB",  "#FC4E07"),
             addEllipses = TRUE, # Concentration ellipses
             ellipse.type = "confidence",
             legend.title = "Clusters",
             repel = TRUE
             )
```
